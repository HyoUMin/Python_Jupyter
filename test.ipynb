````python
# 1. 다음 학생들의 성적 정보를 딕셔너리로 저장하시오

# # 학생1: 이름=김민수, 국어=85, 영어=90, 수학=78

# # 학생2: 이름=이영희, 국어=92, 영어=88, 수학=95

# # 학생3: 이름=박철수, 국어=78, 영어=85, 수학=82

# 2. 각 학생의 총점과 평균을 계산하여 딕셔너리에 추가하시오

student = {
    "student1" : {"이름":"김민수", "국어": 85, "영어": 90, "수학": 78},
    "student2" : {"이름":"이영희", "국어": 92, "영어": 88, "수학": 95},
    "student3" : {"이름":"박철수", "국어": 78, "영어": 85, "수학": 82},
}

stdunt1_total = student["student1"]["국어"] + student["student1"]["영어"] + student["student1"]["수학"]
student1_avg = round(stdunt1_total / 3)

stdunt2_total = student["student2"]["국어"] + student["student2"]["영어"] + student["student2"]["수학"]
student2_avg = round(stdunt1_total / 3)

stdunt3_total = student["student3"]["국어"] + student["student3"]["영어"] + student["student3"]["수학"]
student3_avg = round(stdunt3_total / 3)


student["student1"].update({"총점":stdunt1_total, "평균":student1_avg})
student["student2"].update({"총점":stdunt2_total, "평균":student2_avg})
student["student3"].update({"총점":stdunt3_total, "평균":student3_avg})
````

````python
# 문제 3
# 연도를 입력받아 윤년인지 아닌지 판별하세요.

# 윤년 조건:

# 1. 4로 나누어떨어지는 해는 윤년
# 2. 단, 100으로 나누어떨어지는 해는 평년
# 3. 단, 400으로 나누어떨어지는 해는 윤년

# year = int(input("연도를 입력하세요: "))

year = int(input("연도를 입력하세요: "))

if (year & 4 == 0 and year % 100 != 0) or year % 400 == 0:
    print(f"{year}년은 윤년입니다.")
else:
    print(f"{year}년은 평년입니다.")
````
