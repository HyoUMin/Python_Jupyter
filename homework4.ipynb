```python
# ### 문제 2-2: 계절 판별

# 월(month)을 입력받아 계절을 출력하세요:

# - 3~5월: 봄
# - 6~8월: 여름
# - 9~11월: 가을
# - 12~2월: 겨울
# month = int(input("월을 입력하세요 (1-12): "))

month = int(input("월을 입력하세요 (1-12): "))
match month:
    case 3 | 4 | 5:
        print("3~5월: 봄")
    case 6 | 7 | 8:
        print("6~8월: 여름")
    case 9 | 10 | 11:
        print("9~11월: 가을")
    case 12 | 1 | 2:
        print("12~2월: 겨울")  
```

```python
### 문제 7-1: 택시 요금 계산기

# 거리(km)와 시간대를 입력받아 요금을 계산하세요.

# - 기본요금: 3000원 (2km까지)
# - 추가요금: 2km 초과시 1km당 1000원
# - 심야할증(22시~04시): 총 요금의 20% 추가 
# distance = float(input("거리를 입력하세요 (km): "))
# hour = int(input("현재 시간을 입력하세요 (0-23): "))
# **예상 출력:**
# 거리를 입력하세요 (km): 5
# 현재 시간을 입력하세요 (0-23): 23
# 요금: 7200원

distance = float(input("거리를 입력하세요 (km): "))
hour = int(input("현재 시간을 입력하세요 (0-23): "))
fee = 3000

if  distance > 2:
    fee = fee + ((distance - 2) * 1000)

if hour <= 4 or hour >= 22:
    fee = fee + (fee * 0.2)

print(f"요금: {int(fee)}원")

```

```python
### 문제 7-3: BMI 계산기

# 키와 몸무게를 입력받아 BMI를 계산하고 상태를 출력하세요.

# - BMI = 몸무게(kg) / (키(m) \*\* 2)
# - 18.5 미만: 저체중
# - 18.5 이상 23 미만: 정상
# - 23 이상 25 미만: 과체중
# - 25 이상: 비만

# height = float(input("키를 입력하세요 (cm): "))
# weight = float(input("몸무게를 입력하세요 (kg): "))

# 키를 미터로 변환

# height_m = height / 100

height = float(input("키를 입력하세요 (cm): "))
weight = float(input("몸무게를 입력하세요 (kg): "))

height_m = height / 100
bmi = weight / (height_m ** 2)

if bmi < 18.5:
    print("저체중")
elif bmi < 23:
    print("정상")
elif bmi < 25:
    print("과체중")
else:
    print("비만")

```

```python
## 보너스 문제(선택)

### 윤년 판별기

# 연도를 입력받아 윤년인지 아닌지 판별하세요.

# 윤년 조건:

# 1. 4로 나누어떨어지는 해는 윤년
# 2. 단, 100으로 나누어떨어지는 해는 평년
# 3. 단, 400으로 나누어떨어지는 해는 윤년

# year = int(input("연도를 입력하세요: "))
# **예시:**

# - 2020년: 윤년 (4로 나누어떨어짐)
# - 2100년: 평년 (100으로 나누어떨어지지만 400으로는 안됨)
# - 2000년: 윤년 (400으로 나누어떨어짐)

year = int(input("연도를 입력하세요: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year}년은 윤년입니다.")
else:
    print(f"{year}년은 평년입니다.")
```
