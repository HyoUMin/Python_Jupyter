```python
fruits = {
    'apple': 1000,
    'banana': 1500,
    'orange': 2000,
    'grape': 3000
}

# 1. 모든 과일 이름(key)을 리스트로 출력하시오
print(fruits.keys())

# 2. 모든 가격(value)을 리스트로 출력하시오
print(fruits.values())

# 3. 'banana'의 가격을 출력하시오
print(fruits["banana"])

# 4. 'melon'이 딕셔너리에 있는지 확인하시오
print("melon" in fruits)

# 5. 'grape'를 pop() 함수를 사용하여 삭제하고, 삭제된 값을 출력하시오
print(fruits.pop("grape"))
```

```python
# 학생1: 이름=김민수, 국어=85, 영어=90, 수학=78
# 학생2: 이름=이영희, 국어=92, 영어=88, 수학=95
# 학생3: 이름=박철수, 국어=78, 영어=85, 수학=82
s1 = {'이름': '김민수', '국어': 85, '영어': 90, '수학': 78}
s2 = {'이름': '이영희', '국어': 92, '영어': 88, '수학': 95}
s3 = {'이름': '박철수', '국어': 78, '영어': 85, '수학': 82}

s1['총점'] = s1['국어'] + s1['영어'] + s1['수학']
s1['평균'] = s1['총점'] / 3
a1 = s1['총점']
b1 = s1['평균']

s2['총점'] = s2['국어'] + s2['영어'] + s2['수학']
s2['평균'] = s2['총점'] / 3
a2 = s2['총점']
b2 = s2['평균']

s3['총점'] = s3['국어'] + s3['영어'] + s3['수학']
s3['평균'] = s3['총점'] / 3
a3 = s3['총점']
b3 = s3['평균']

print("=== 학생 성적 관리 ===")
print("이름:", s1['이름'])
print(f"국어: {s1['국어']}점, 영어: {s1['영어']}점, 수학: {s1['수학']}점")
print("총점:", f"{a1}점", "평균:", f"{b1:.2f}점")

print("---")

print("이름:", s2['이름'])
print(f"국어: {s2['국어']}점, 영어: {s2['영어']}점, 수학: {s2['수학']}점")
print("총점:", f"{a2}점", "평균:", f"{b2:.2f}점")

print("---")

print("이름:", s3['이름'])
print(f"국어: {s3['국어']}점, 영어: {s3['영어']}점, 수학: {s3['수학']}점")
print("총점:", f"{a3}점", "평균:", f"{b3:.2f}점")

print("---")
```

```python
a_class = {'김철수', '이영희', '박민수', '최지영', '홍길동'}
b_class = {'박민수', '최지영', '정수민', '강호동', '이영희'}




# 1. 두 반 모두에 있는 학생 (교집합)
a = a_class.intersection(b_class) 
print(a)
# 2. 두 반을 합친 전체 학생 (합집합)
b = a_class.union(b_class) 
print(b)

# 3. A반에만 있는 학생 (차집합)
c = a_class.difference(b_class)
print(c)
 
# 4. B반에만 있는 학생 (차집합)
d = b_class.difference(a_class)
print(d)
```

```python
# 문제 12. 집합 조작하기

fruits = {'apple', 'banana', 'orange'}


# 다음 작업을 수행하는 코드를 작성하시오:


# 1. 'grape'를 추가하시오
fruits.add('grape')
print(fruits)
'orange' in fruits
# 2. 'melon', 'kiwi', 'mango'를 한 번에 추가하시오
fruits.update(("melon", "kiwi", "mango"))
print(fruits)
# 3. 'banana'를 제거하시오
fruits.remove("banana")
print(fruits)

# 4. 집합에 'strawberry'가 있는지 확인하시오
a = 'strawberry' in fruits
print(a)
# 5. 모든 과일을 제거하고 빈 집합으로 만드시오
fruits.clear()
print(fruits)
```

```python
#문제 8. 단어 카운터
#문자열에서 각 단어가 몇 번 등장하는지 세는 프로그램을 작성하시오.

text = "python is great python is fun python is powerful"

# 요구사항:
# 1. 문자열을 공백으로 분리하여 단어 리스트 만들기
# 2. 각 단어의 등장 횟수 세기
# 3. 결과를 출력


words = list(text.split())

print(f"python: {words.count('python')}번")
print(f"is: {words.count('is')}번")
print(f"great: {words.count('great')}번")
print(f"fun: {words.count('fun')}번")
print(f"powerful {words.count('powerful')}번")

#**출력 예시:**

#python: 3번
#is: 3번
#great: 1번
#fun: 1번
#powerful: 1번
```
